```python

# Student Profile System

print("===== STUDENT PROFILE SYSTEM =====")

# Input
name = input("Enter your full name: ")

age_input = input("Enter your age: ")
age = int(age_input)

gpa_input = input("Enter your GPA: ")
gpa = float(gpa_input)

courses_input = input("Enter courses (separated by commas): ")
courses = courses_input.split(",")

# Remove extra spaces from course names
courses = [course.strip() for course in courses]

# Boolean values
deans_list = gpa >= 3.5
is_adult = age >= 18

# Dictionary
student_profile = {
    "name": name,
    "age": age,
    "gpa": gpa,
    "courses": courses,
    "num_courses": len(courses)
}

# Type Conversion Demonstration
print("\n===== TYPE CONVERSION =====")
print(f"Age Input Type: {type(age_input)}")
print(f"Age Converted Type: {type(age)}")

print(f"GPA Input Type: {type(gpa_input)}")
print(f"GPA Converted Type: {type(gpa)}")

print(f"Courses Input Type: {type(courses_input)}")
print(f"Courses Converted Type: {type(courses)}")

# Display Profile
print("\n===== STUDENT PROFILE =====")
print(f"Name:        {student_profile['name']}")
print(f"Age:         {student_profile['age']} ({'Adult' if is_adult else 'Minor'})")
print(f"GPA:         {student_profile['gpa']}")
print(f"Courses:     {student_profile['courses']}")
print(f"# Courses:   {student_profile['num_courses']}")
print(f"Dean's List: {'Yes' if deans_list else 'No'}")

# Data Types Demonstration
print("\n===== DATA TYPES USED =====")
print(type(name))              # str
print(type(age))               # int
print(type(gpa))               # float
print(type(deans_list))        # bool
print(type(courses))           # list
print(type(student_profile))   # dict
```
